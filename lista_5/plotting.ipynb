```julia
using LinearAlgebra
dane = ["./Dane/Dane16_1_1", "./Dane/Dane10000_1_1", "./Dane/Dane50000_1_1", "./Dane/Dane100000_1_1", "./Dane/Dane300000_1_1", "./Dane/Dane500000_1_1"]

err = []
tim = []

for dana in dane 
    a_path = "$(dana)/A.txt"
    b_path = "$(dana)/b.txt"
    println(dana)
    a = @time FastAdaptedMatrix(a_path)
    b = Vector(b_path)
    n = a.n
    l = a.l
    t1 = @elapsed a = LU!(a)
    t2 = @elapsed x = LUxb!(a,b)
    push!(err, norm(x - ones(n)) / norm(ones(n)))
    push!(tim, t1 + t2)
end

println(err)
println(tim)
```

```julia
using LinearAlgebra
dane = ["./Dane/Dane16_1_1", "./Dane/Dane10000_1_1", "./Dane/Dane50000_1_1", "./Dane/Dane100000_1_1", "./Dane/Dane300000_1_1", "./Dane/Dane500000_1_1"]

err_p = []
time_p = []

for dana in dane 
    a_path = "$(dana)/A.txt"
    b_path = "$(dana)/b.txt"
    println(dana)
    a = @time FastAdaptedMatrix(a_path)
    b = Vector(b_path)
    n = a.n
    l = a.l
    t1 = @elapsed a,p = pivotedLU!(a)
    t2 = @elapsed x = pivotedLUxb!(a,p,b)
    push!(err_p, norm(x - ones(n)) / norm(ones(n)))
    push!(time_p, t1 + t2)
end

println(err_p)
println(time_p)
```

```julia
n = [16, 10000, 50000, 100000, 300000, 500000]
l = [4 , 5    , 4    , 5     , 4     , 4]
cases = 6

using Printf
fix_e(x, dig = 4) = replace(string(round(x, sigdigits = dig)), r"e\+00" => "", r"e(-\d+)" => s" dot 10^(\1)", r"e\+?(\d+)" => s" dot 10^(\1)")

print("\$n,l\$,")
for i in 1:cases
    print("\$$(n[i]),$(l[i])\$,")
end
println()
print("[Gauss],")
for i in 1:cases
    t = fix_e(time_g[i])
    print("\$$t s\$,")
end
println()
print("[Gauss z wyb.],")
for i in 1:cases
    t = fix_e(time_gp[i])
    print("\$$t s\$,")
end
println()
print("[LU],")
for i in 1:cases
    t = fix_e(tim[i])
    print("\$$t s\$,")
end
println()
print("[LU z wyb.],")
for i in 1:cases
    t = fix_e(time_p[i])
    print("\$$t s\$,")
end
println()
```
